## Load the dataset

### Subtask:
Load the `Train_set.csv` file into a pandas DataFrame.


**Reasoning**:
Import the pandas library and load the `Train_set.csv` file into a pandas DataFrame and display the head of the dataframe to confirm successful loading.



```python
import pandas as pd

df = pd.read_csv('/content/Train_set.csv')
display(df.head())
```

## Preprocess the data

### Subtask:
Clean and prepare the text data for model training (e.g., handle missing values, tokenize).


**Reasoning**:
Check for missing values in the '문장' column and remove leading/trailing whitespace from the text data.



```python
print("Missing values before cleaning:")
print(df['문장'].isnull().sum())

df['문장'] = df['문장'].str.strip()

print("\nMissing values after cleaning:")
print(df['문장'].isnull().sum())
display(df.head())
```

## Split the dataset

### Subtask:
Split the data into stratified training and testing sets.


**Reasoning**:
Split the DataFrame into stratified training and testing sets using train_test_split, ensuring the split is stratified based on the 'label' column.



```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['label'])

print("Training set shape:", train_df.shape)
print("Testing set shape:", test_df.shape)
display(train_df.head())
display(test_df.head())
```

## Load the pre-trained model

### Subtask:
Load the KcBERT model and tokenizer using the transformers library.


**Reasoning**:
Import the necessary classes and load the tokenizer and model for KcBERT.



```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

model_name = "beomi/kcbert-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=6) # Assuming 6 labels based on previous steps
```

## Prepare data for training

### Subtask:
Tokenize and format the data for input into the model.


**Reasoning**:
Define a function to tokenize the text and encode the labels, then apply it to the training and testing DataFrames and convert them to Hugging Face Dataset objects.



```python
from datasets import Dataset

def tokenize_function(row):
    inputs = tokenizer(row['문장'], padding='max_length', truncation=True)
    inputs['labels'] = row['label']
    return inputs

train_dataset = Dataset.from_pandas(train_df)
test_dataset = Dataset.from_pandas(test_df)

tokenized_train_dataset = train_dataset.map(tokenize_function, batched=True)
tokenized_test_dataset = test_dataset.map(tokenize_function, batched=True)

print("Tokenized training dataset:")
print(tokenized_train_dataset)
print("\nTokenized testing dataset:")
print(tokenized_test_dataset)
```

## Fine-tune the model

### Subtask:
Train the KcBERT model on the training set.


**Reasoning**:
Train the KcBERT model on the training set using the Hugging Face Trainer.



**Reasoning**:
The previous command failed because `evaluation_strategy` is not a valid argument for `TrainingArguments` in the current version of the transformers library. The argument should be `eval_strategy`. I will correct the argument name and rerun the training code.



```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
    eval_strategy="epoch",  # Corrected argument name
    report_to="none", # Disable W&B integration
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train_dataset,
    eval_dataset=tokenized_test_dataset,
)

trainer.train()
```

## Handle duplicates

### Subtask:
Remove duplicate sentences from the DataFrame.


**Reasoning**:
Remove duplicate rows based on the '문장' column and print the shape before and after removing duplicates.



```python
print("Shape before removing duplicates:", reviews_df.shape)
reviews_df.drop_duplicates(subset=['문장'], inplace=True)
print("Shape after removing duplicates:", reviews_df.shape)
display(reviews_df.head())
```

## Remove special characters

### Subtask:
Clean the text data by removing special characters.


**Reasoning**:
Define a function to remove special characters and apply it to the '문장' column.



```python
import re

def remove_special_characters(text):
    text = re.sub(r'[^가-힣a-zA-Z0-9\s]', '', text)
    return text

reviews_df['문장'] = reviews_df['문장'].apply(remove_special_characters)
display(reviews_df.head())
```

## Filter long sentences

### Subtask:
Remove sentences that exceed a certain length threshold.


**Reasoning**:
Define a function to filter sentences by length, apply it to the DataFrame, filter the DataFrame, print shapes, and display the head of the filtered DataFrame.



```python
def filter_sentence_length(sentence, max_length=200):
    return len(sentence) <= max_length

print("Shape before filtering by length:", reviews_df.shape)

length_mask = reviews_df['문장'].apply(filter_sentence_length)
reviews_df = reviews_df[length_mask]

print("Shape after filtering by length:", reviews_df.shape)
display(reviews_df.head())
```

```python
import pandas as pd

preprocessed_reviews_df = pd.read_excel('/content/preprocessed_reviews.xlsx')
display(preprocessed_reviews_df.head())
preprocessed_reviews_df.info()
```

## Load the fine-tuned model and tokenizer

### Subtask:
Load the fine-tuned KcBERT model and tokenizer from the specified local path.


**Reasoning**:
Import the necessary classes and load the tokenizer and model for KcBERT from the specified local path.



```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from google.colab import drive
import os

# Mount Google Drive if not already mounted
if not os.path.exists('/content/Drive'):
    drive.mount('/content/Drive')
else:
    print("Google Drive already mounted.")


model_path = "/content/fine_tuned_kcbert"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSequenceClassification.from_pretrained(model_path)
```

```python
def tokenize_sentence(sentence):
    inputs = tokenizer(sentence, padding=True, truncation=True, return_tensors="pt")
    return inputs

tokenized_reviews = preprocessed_reviews_df['문장'].apply(tokenize_sentence)
```

```python
import torch

def predict_sentiment(tokenized_input):
    with torch.no_grad():
        outputs = model(**tokenized_input)
    logits = outputs.logits
    probabilities = torch.softmax(logits, dim=1).squeeze().tolist()
    return probabilities

sentiment_probabilities = tokenized_reviews.apply(predict_sentiment)
```

```python
def is_probability_similar(probabilities, threshold=0.1):
    sorted_probs = sorted(probabilities, reverse=True)
    return (sorted_probs[0] - sorted_probs[1]) < threshold

similar_probabilities_mask = sentiment_probabilities.apply(is_probability_similar)

print("Number of sentences with similar probabilities:", similar_probabilities_mask.sum())
```

```python
similar_probability_sentences_df = preprocessed_reviews_df[similar_probabilities_mask].copy()
similar_probability_sentences_df['softmax_probabilities'] = sentiment_probabilities[similar_probabilities_mask]

similar_probability_sentences_df.to_excel('similar_probability_sentences.xlsx', index=False)

print("\nSentences with similar probabilities extracted and saved to 'similar_probability_sentences.xlsx'")
display(similar_probability_sentences_df.head())
```

```python
preprocessed_reviews_df['softmax_probabilities'] = sentiment_probabilities
# Assuming you want to add predicted labels as well, though not explicitly in the plan step
# You would need to get the predicted labels from the softmax_probabilities if needed
# For now, just saving probabilities as requested by the overall task

preprocessed_reviews_df.to_excel('reviews_with_probabilities.xlsx', index=False)

print("\nOriginal reviews with softmax probabilities saved to 'reviews_with_probabilities.xlsx'")
```
